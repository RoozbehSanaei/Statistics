```python
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
import numpy as np
#
# Instance of Bernoulli distribution with parameter p = 0.7
#
#
# Outcome of experiment can take value as 0, 1
#
#
# Create a bar plot; Note the usage of "pmf" function
# to determine the probability of different values of
# random variable
#
def plot_pmf(f,range):
  fig, ax = plt.subplots(1, 1)
  fig.set_size_inches(18.5, 10.5)
  ax.plot(np.arange(range[0], range[1], 1), f(np.arange(range[0], range[1], 1)), 'bo', ms=8, label='bernoulli pmf')
  ax.vlines(np.arange(range[0], range[1], 1), 0, f(np.arange(range[0], range[1], 1)), colors='r', lw=5, alpha=0.5)
  ax.xaxis.set_ticks(np.arange(-10, 10, 1))
  ax.set_xlabel("x",fontsize=18)
  ax.set_ylabel("P(X=x)",fontsize=18)
  ax.tick_params(axis='x', labelsize=16)
  ax.tick_params(axis='y', labelsize=16)
  plt.show()



def plot_cdf(f,range):
  fig, ax = plt.subplots(1, 1)
  fig.set_size_inches(18.5, 10.5)
  ax.plot(np.arange(range[0], range[1], 0.01),f(np.arange(range[0], range[1], 0.01)),'b')
  ax.plot(np.arange(range[0], range[1], 1),f(np.arange(range[0], range[1], 1)),'ro')
  ax.xaxis.set_ticks(np.arange(-10, 10, 1))
  ax.set_xlabel("x",fontsize=18)
  ax.set_ylabel("Culti",fontsize=18)
  ax.tick_params(axis='x', labelsize=16)
  ax.tick_params(axis='y', labelsize=16)
  plt.show()
```

```python
def f(x): return bernoulli.cdf(x,0.3)

plot_pmf(f,(-10,10))
plt.show()
```

```python

def f(x): return bernoulli.cdf(x,0.3)

plot_cdf(f,(-10,10))
plt.show()
```
