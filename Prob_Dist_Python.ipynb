```python
import matplotlib.pyplot as plt
from scipy.stats import bernoulli,binom,geom
import numpy as np
import numpy.ma as ma

#
# Instance of Bernoulli distribution with parameter p = 0.7
#
#
# Outcome of experiment can take value as 0, 1
#
#
# Create a bar plot; Note the usage of "pmf" function
# to determine the probability of different values of
# random variable
#
def plot_pmf(f,range):
  fig, ax = plt.subplots(1, 1)
  fig.set_size_inches(18.5, 10.5)
  ax.plot(np.arange(range[0], range[1], 1), f(np.arange(range[0], range[1], 1)), 'bo', ms=8, label='bernoulli pmf')
  ax.vlines(np.arange(range[0], range[1], 1), 0, f(np.arange(range[0], range[1], 1)), colors='r', lw=5, alpha=0.5)
  ax.xaxis.set_ticks(np.arange(-10, 10, 1))
  ax.set_xlabel("x",fontsize=36)
  ax.set_ylabel("P(X=x)",fontsize=36)
  ax.tick_params(axis='x', labelsize=32,rotation=90)
  ax.tick_params(axis='y', labelsize=32)
  ax.set_ylim([-0.1,1.1])
  plt.show()



def plot_cdf(f,r):

  fig, ax = plt.subplots(1, 1)
  fig.set_size_inches(18.5, 10.5)
  mc = ma.array(f(np.arange(r[0], r[1], 0.01)))
  for j in range(0,mc.shape[0],100):
    mc[j] = ma.masked
  ax.plot(np.arange(r[0], r[1], 0.01),mc,'b')
  ax.plot(np.arange(r[0], r[1], 1),f(np.arange(r[0], r[1], 1)),'ro')
  ax.xaxis.set_ticks(np.arange(-10, 10, 1))
  ax.set_xlabel("x",fontsize=36)
  ax.set_ylabel("P(X<=x)",fontsize=36)
  ax.tick_params(axis='x', labelsize=32,rotation=90)
  ax.tick_params(axis='y', labelsize=32)
  ax.set_ylim([-0.1,1.1])
  plt.show()
```

```python
def f(x): return bernoulli.pmf(x,0.3)

plot_pmf(f,(-10,10))
plt.show()
```

```python

def f(x): return bernoulli.cdf(x,0.3)

plot_cdf(f,(-10,10))
plt.show()
```

```python
def f(x): return binom.pmf(x,10,0.3)

plot_pmf(f,(-10,10))
plt.show()
```

```python

def f(x): return geom.pmf(x,0.4)

plot_pmf(f,(-10,10))
plt.show()
```

```python
def f(x): return geom.cdf(x,0.4)

plot_cdf(f,(-10,10))
plt.show()
```
